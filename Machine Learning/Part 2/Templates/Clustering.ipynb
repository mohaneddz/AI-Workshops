# Template - Clustering
Covers `KMeans`, `DBSCAN`, and `Agglomerative` with dendrogram.


```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, dendrogram

sns.set_theme(style="whitegrid")
DATA_PATH = Path("../Solutions/Clustering/data.csv")
df = pd.read_csv(DATA_PATH)
X = df[["age", "annual_income_k", "spending_score"]].copy()
X_scaled = StandardScaler().fit_transform(X)
X.head()
```

## 1) KMeans


```python
kmeans = KMeans(n_clusters=5, random_state=42, n_init=20)
df["kmeans_cluster"] = kmeans.fit_predict(X_scaled)

plt.figure(figsize=(7, 5))
sns.scatterplot(data=df, x="annual_income_k", y="spending_score", hue="kmeans_cluster", palette="tab10", s=70)
plt.title("KMeans Clusters")
plt.tight_layout()
plt.show()
```

## 2) DBSCAN


```python
db = DBSCAN(eps=0.8, min_samples=6)
df["dbscan_cluster"] = db.fit_predict(X_scaled)

plt.figure(figsize=(7, 5))
sns.scatterplot(data=df, x="annual_income_k", y="spending_score", hue="dbscan_cluster", palette="tab10", s=70)
plt.title("DBSCAN Clusters (-1 = noise)")
plt.tight_layout()
plt.show()
```

## 3) Agglomerative + Dendrogram


```python
agg = AgglomerativeClustering(n_clusters=5, linkage="ward")
df["agg_cluster"] = agg.fit_predict(X_scaled)

plt.figure(figsize=(7, 5))
sns.scatterplot(data=df, x="annual_income_k", y="spending_score", hue="agg_cluster", palette="Set2", s=70)
plt.title("Agglomerative Clusters")
plt.tight_layout()
plt.show()

linked = linkage(X_scaled, method="ward")
plt.figure(figsize=(10, 5))
dendrogram(linked, truncate_mode="lastp", p=20, show_leaf_counts=True)
plt.title("Dendrogram (Agglomerative)")
plt.tight_layout()
plt.show()
```
