# Template - Dimensionality Reduction
Compares `PCA`, `t-SNE`, and `UMAP` on the same data.


```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap.umap_ as umap

sns.set_theme(style="whitegrid")
DATA_PATH = Path("../Solutions/Dimensionality Reduction/data.csv")
df = pd.read_csv(DATA_PATH)
X = df[["sepal_length_cm", "sepal_width_cm", "petal_length_cm", "petal_width_cm"]]
y = df["species"]
X_scaled = StandardScaler().fit_transform(X)
```

## 1) PCA


```python
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)
print("Explained variance ratio:", pca.explained_variance_ratio_.round(3))

pca_df = pd.DataFrame({"x": X_pca[:,0], "y": X_pca[:,1], "species": y})
plt.figure(figsize=(7, 5))
sns.scatterplot(data=pca_df, x="x", y="y", hue="species", s=80)
plt.title("PCA Projection")
plt.tight_layout()
plt.show()
```

## 2) t-SNE


```python
tsne = TSNE(n_components=2, random_state=42, perplexity=20)
X_tsne = tsne.fit_transform(X_scaled)
tsne_df = pd.DataFrame({"x": X_tsne[:,0], "y": X_tsne[:,1], "species": y})

plt.figure(figsize=(7, 5))
sns.scatterplot(data=tsne_df, x="x", y="y", hue="species", s=80)
plt.title("t-SNE Projection")
plt.tight_layout()
plt.show()
```

## 3) UMAP


```python
um = umap.UMAP(n_components=2, random_state=42, n_neighbors=15, min_dist=0.2)
X_umap = um.fit_transform(X_scaled)
umap_df = pd.DataFrame({"x": X_umap[:,0], "y": X_umap[:,1], "species": y})

plt.figure(figsize=(7, 5))
sns.scatterplot(data=umap_df, x="x", y="y", hue="species", s=80)
plt.title("UMAP Projection")
plt.tight_layout()
plt.show()
```
